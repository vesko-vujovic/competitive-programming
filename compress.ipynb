```python
A = 'ABBBCCDDCCC'
K = 3
# A = 'AAAAAAAAAAABXXAAAAAAAAAA'
```

```python
def solution(A, K):
    count = 0
    pos  = 0
    n = len(A) - 1
    compressed = ''

    while pos <= n:
        
        curr = A[pos]
        num  = 0

        while  pos <= n and ( A[pos] == curr ):

            pos = pos + 1
            num = num + 1

            k = pos + K 

            if(k <= n) and (curr != A[pos]) and (curr == A[k]):
                pos = k 

        
        if num != 1: 
            compressed = compressed + str(num)
        compressed = compressed + curr

    return compressed
            

```

```python
solution = solution(A, K)
```
